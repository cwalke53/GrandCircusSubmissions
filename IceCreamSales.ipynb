## Question 2:In addition to the number of sales, the ice-cream shop also records its daily revenue ($ in sales). Theyâ€™ have also captured the noon temperature for the last 10 days: Create a scatter plot to represent the data. What can you conclude from the scatterplot? Make sure you have the appropriate labels for the x and y axis......As the temperature rises the sales increase hinting a positive correlation........





```python
df = pd.DataFrame({
   'Temperature (F)': [57.56, 61.52, 53.42, 59.36, 65.3, 71.78, 66.92, 77.18, 74.12, 64.58],
   'Sales': [215, 325, 185, 332, 406, 522, 412, 614, 544, 421]
})

```

```python
plt.scatter(df['Sales'], df['Temperature (F)'], c="#00f", marker="*",s=150, alpha=0.3)
plt.title('Frozen Yogurt Sales vs Temperature')
plt.xlabel('Amount of Sales')
plt.ylabel('Temperature')
```
