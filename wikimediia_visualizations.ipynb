```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
df = pd.read_csv('combined_output.csv')
```

```python
df['bot'] = df['bot'].astype(bool)
df.info(# Column Chart
plt.figure(figsize=(8, 6))
plt.bar(df['Category'], df['Values1'], color='skyblue', label='Values1')
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Column Chart for Values1')
plt.legend()
plt.show()
)
```

```python
df.info()
```

```python
# Histogram and KDE
plt.figure(figsize=(8, 6))
sns.histplot(df['length_diff'], kde=True, bins=30, color='skyblue')
plt.title("Distribution of Length Difference")
plt.xlabel("Length Difference")
plt.ylabel("Frequency")
plt.show()
```

```python
# Proportion of Bots
bot_counts = df['bot'].value_counts()

# Pie Chart
plt.figure(figsize=(6, 6))
bot_counts.plot(kind='pie', autopct='%1.1f%%', labels=['Human', 'Bot'], colors=['lightgreen', 'orange'], startangle=90)
plt.title("Proportion of Bot vs Human Edits")
plt.ylabel("")  # Remove y-label for aesthetics
plt.show()
```

```python
# Top 10 Domains with Most Edits
top_domains = df['domain'].value_counts().head(10)

# Horizontal Bar Chart
plt.figure(figsize=(10, 6))
top_domains.plot(kind='barh', color='skyblue')
plt.title("Top 10 Domains by Number of Edits")
plt.xlabel("Number of Edits")
plt.ylabel("Domain")
plt.gca().invert_yaxis()  # Invert to show highest first
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['old_length', 'new_length', 'length_diff']].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap of Length Metrics")
plt.show()
```
