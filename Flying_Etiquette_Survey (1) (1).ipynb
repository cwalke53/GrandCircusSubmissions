### 1) As part of data preparation, use the information from "How tall are you?" to produce a new "Height (Inches)" column. Begin by creating the “Height (inches)” column as a copy of the “How tall are you?” column.


```python
df["Height (Inches)"] = df.loc[:, "How tall are you?"]
df
```

### A) Change Under 5 ft. to 5’0”

```python
df[ "Height (Inches)"] = df["Height (Inches)"].replace(["Under 5 ft.", "5'0"])
```

### B) Change 6’6” and above to 6’6”

```python
df["Height (Inches)"] = df["Height (Inches)"].replace(['6\'6" and above', "6'6"])
df["Height (Inches)"].unique()
```

```python
df["Height (Inches)"] = df["Height (Inches)"].fillna("0'0")
df["Height (Inches)"].unique()
```

### Remove extra '"' in height(inches)

```python
df["Height (Inches)"] = df["Height (Inches)"].str.replace( '"', '').str.strip()
df["Height (Inches)"].unique()
```

### C) Replace the “Height (Inches)” column, converting the feet and inches to an integer of total inches.


```python
df["Height (Inches)"] = (df["Height (Inches)"].str.split("'").str[0].astype(int) * 12) + (
    df["Height (Inches)"].str.split("'").str[1].astype(int)
)
```
